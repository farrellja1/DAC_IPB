# Data Import
Let's import all the test data from the different models (Claude, DeepSeek, Gemini, GPT, Grok, and Perplexity) stored in the Test directory.

```python
import pandas as pd
import os

# Path to the Test directory
test_path = "data/Test"

# Dictionary to store all test dataframes
test_data = {}

# Read all CSV files from the Test directory
for file in os.listdir(test_path):
    if file.endswith('.csv'):
        model_name = file.split('.')[0]  # Get model name without extension
        file_path = os.path.join(test_path, file)
        test_data[model_name] = pd.read_csv(file_path)
        print(f"Loaded {model_name} test data with shape: {test_data[model_name].shape}")

# Display first few rows of each dataset
for model_name, df in test_data.items():
    print(f"\nFirst few rows of {model_name} test data:")
    display(df.head())
```

# Training Data Import
Now let's import all the training data from the different models in the Train directory.

```python
# Path to the Train directory
train_path = "data/Train"

# Dictionary to store all training dataframes
train_data = {}

# Read all CSV files from the Train directory
for file in os.listdir(train_path):
    if file.endswith('.csv'):
        model_name = file.split('.')[0]  # Get model name without extension
        file_path = os.path.join(train_path, file)
        train_data[model_name] = pd.read_csv(file_path)
        print(f"Loaded {model_name} training data with shape: {train_data[model_name].shape}")

# Display first few rows of each dataset
for model_name, df in train_data.items():
    print(f"\nFirst few rows of {model_name} training data:")
    display(df.head())
```

# Version Analysis
Let's examine all the unique versions for each model in both training and test datasets.

```python
def analyze_versions(data_dict, dataset_name):
    print(f"\n{dataset_name} Dataset Version Analysis:")
    print("=" * 50)
    
    for model_name, df in data_dict.items():
        # Convert versions to strings
        df['AppVersion'] = df['AppVersion'].astype(str)
        versions = sorted(df['AppVersion'].unique())
        version_counts = df['AppVersion'].value_counts().sort_index()
        
        print(f"\n{model_name}:")
        print(f"Number of unique versions: {len(versions)}")
        print("Version distribution:")
        for version, count in version_counts.items():
            percentage = round((count / len(df) * 100), 2)
            print(f"  {version}: {count} comments ({percentage}%)")
        print(f"First version: {versions[0]}")
        print(f"Latest version: {versions[-1]}")
        print("-" * 30)

# Analyze versions in training data
analyze_versions(train_data, "Training")

# Analyze versions in test data
analyze_versions(test_data, "Test")
```

# Text Preprocessing and Word Clouds
Let's analyze the most common words in comments for each model using word clouds. We'll follow these preprocessing steps:
1. Lowercase conversion
2. Punctuation and special character removal
3. Number removal
4. Tokenization
5. Stopword removal
6. Lemmatization
7. Short token removal

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import re

# Download required NLTK data
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove special characters and numbers, keeping only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return ' '.join(tokens)
```

```python
import matplotlib.pyplot as plt

# Process test and train data together for each model
models = ['claude', 'deepseek', 'gemini', 'gpt', 'grok', 'perplexity']
fig = plt.figure(figsize=(20, 10))

for idx, model in enumerate(models, 1):
    # Combine test and train comments
    test_comments = ' '.join(test_data[model]['Comment'].astype(str))
    train_comments = ' '.join(train_data[model]['Comment'].astype(str))
    all_comments = test_comments + ' ' + train_comments
    
    # Preprocess text
    processed_text = preprocess_text(all_comments)
    
    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400,
                         background_color='white',
                         min_font_size=10,
                         max_font_size=150,
                         colormap='viridis').generate(processed_text)
    
    # Add subplot
    plt.subplot(2, 3, idx)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'{model.capitalize()} Word Cloud')

plt.tight_layout(pad=3.0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

# Download required NLTK data (run once)
nltk.download('punkt_tab', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

def preprocess_text(text):
    """
    Preprocess text by cleaning, tokenizing, and lemmatizing
    """
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove special characters and numbers, keeping only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    # Add common app review words that aren't very meaningful
    additional_stops = {'app', 'application', 'use', 'using', 'get', 'one', 'would', 'could', 'also', 'really'}
    stop_words.update(additional_stops)
    tokens = [token for token in tokens if token not in stop_words and len(token) > 2]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return ' '.join(tokens)

def create_negative_wordclouds(train_data):
    """
    Create word clouds for negative reviews (sentiment = 0) for each model
    """
    models = list(train_data.keys())
    
    # Create figure with subplots
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('Word Clouds for Negative Reviews (Sentiment = 0)', fontsize=20, fontweight='bold')
    
    # Flatten axes for easier iteration
    axes = axes.flatten()
    
    for idx, model in enumerate(models):
        # Filter for negative reviews only (sentiment = 0)
        negative_reviews = train_data[model][train_data[model]['Sentiment'] == 0]
        
        if len(negative_reviews) == 0:
            # Handle case where there are no negative reviews
            axes[idx].text(0.5, 0.5, f'No negative reviews\nfor {model.capitalize()}', 
                          ha='center', va='center', transform=axes[idx].transAxes, 
                          fontsize=16, fontweight='bold')
            axes[idx].set_title(f'{model.capitalize()} (0 negative reviews)', fontsize=14, fontweight='bold')
            axes[idx].axis('off')
            continue
        
        # Combine all negative comments
        negative_text = ' '.join(negative_reviews['Comment'].astype(str))
        
        # Preprocess the text
        processed_text = preprocess_text(negative_text)
        
        if not processed_text.strip():
            # Handle case where processed text is empty
            axes[idx].text(0.5, 0.5, f'No meaningful words\nin negative reviews', 
                          ha='center', va='center', transform=axes[idx].transAxes, 
                          fontsize=16, fontweight='bold')
            axes[idx].set_title(f'{model.capitalize()} ({len(negative_reviews)} negative reviews)', 
                               fontsize=14, fontweight='bold')
            axes[idx].axis('off')
            continue
        
        # Generate word cloud with a red color scheme for negative sentiment
        wordcloud = WordCloud(
            width=800, 
            height=400,
            background_color='white',
            min_font_size=10,
            max_font_size=100,
            colormap='Reds',  # Red color scheme for negative reviews
            max_words=100,
            relative_scaling=0.5,
            stopwords=set(stopwords.words('english'))
        ).generate(processed_text)
        
        # Display the word cloud
        axes[idx].imshow(wordcloud, interpolation='bilinear')
        axes[idx].axis('off')
        axes[idx].set_title(f'{model.capitalize()} ({len(negative_reviews)} negative reviews)', 
                           fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)  # Make room for main title
    plt.show()
    
    # Print summary statistics
    print("\n" + "="*60)
    print("NEGATIVE REVIEWS SUMMARY")
    print("="*60)
    
    for model in models:
        negative_count = len(train_data[model][train_data[model]['Sentiment'] == 0])
        total_count = len(train_data[model])
        percentage = (negative_count / total_count * 100) if total_count > 0 else 0
        
        print(f"{model.capitalize():12} | {negative_count:4d} negative reviews ({percentage:.1f}% of total)")

# Usage example (assuming train_data dictionary is already loaded):
create_negative_wordclouds(train_data)
```

## Language Distribution Analysis

```python
from langdetect import detect
from collections import Counter
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def detect_language(text):
    try:
        return detect(str(text))
    except:
        return 'unknown'

# Function to get language distribution for a dataset
def get_language_dist(data):
    languages = []
    for comment in data['Comment']:
        lang = detect_language(comment)
        languages.append(lang)
    return Counter(languages)

# Get language distribution for each model
models = ['claude', 'deepseek', 'gemini', 'gpt', 'grok', 'perplexity']
test_langs = {}
train_langs = {}

for model in models:
    test_langs[model] = get_language_dist(test_data[model])
    train_langs[model] = get_language_dist(train_data[model])

# Create subplots
fig = make_subplots(rows=2, cols=1, 
                    subplot_titles=('Language Distribution in Test Data',
                                  'Language Distribution in Training Data'))

# Add traces for test data
for model in models:
    dist = test_langs[model]
    langs = list(dist.keys())
    counts = list(dist.values())
    
    fig.add_trace(
        go.Bar(name=model.capitalize(), x=langs, y=counts),
        row=1, col=1
    )

# Add traces for training data
for model in models:
    dist = train_langs[model]
    langs = list(dist.keys())
    counts = list(dist.values())
    
    fig.add_trace(
        go.Bar(name=model.capitalize(), x=langs, y=counts),
        row=2, col=1
    )

# Update layout
fig.update_layout(height=1000,
                 barmode='group',
                 title_text="Language Distribution in Comments")

fig.update_xaxes(title_text="Language", row=1, col=1)
fig.update_xaxes(title_text="Language", row=2, col=1)
fig.update_yaxes(title_text="Number of Comments", row=1, col=1)
fig.update_yaxes(title_text="Number of Comments", row=2, col=1)

fig.show()
```

## Aggregated Language Distribution
Let's look at the total number of comments per language across all models.

```python
# Aggregate language counts across all models
test_total = Counter()
train_total = Counter()

# Sum up the counts for each language
for model in models:
    test_total.update(test_langs[model])
    train_total.update(train_langs[model])

# Create DataFrames for test and training data
test_df = pd.DataFrame.from_dict(test_total, orient='index', columns=['count']).reset_index()
test_df.columns = ['language', 'count']
test_df = test_df.sort_values('count', ascending=False)  # Sort descending to get top languages
test_df = test_df.head(10)  # Keep only top 10
test_df = test_df.sort_values('count', ascending=True)  # Sort ascending for visualization

train_df = pd.DataFrame.from_dict(train_total, orient='index', columns=['count']).reset_index()
train_df.columns = ['language', 'count']
train_df = train_df.sort_values('count', ascending=False)  # Sort descending to get top languages
train_df = train_df.head(10)  # Keep only top 10
train_df = train_df.sort_values('count', ascending=True)  # Sort ascending for visualization

# Create subplots
fig = make_subplots(rows=2, cols=1,
                    subplot_titles=('Top 10 Languages in Test Data (All Models)',
                                  'Top 10 Languages in Training Data (All Models)'),
                    vertical_spacing=0.12)

# Add horizontal bar chart for test data
fig.add_trace(
    go.Bar(y=test_df['language'],
           x=test_df['count'],
           orientation='h',
           name='Test Data'),
    row=1, col=1
)

# Add horizontal bar chart for training data
fig.add_trace(
    go.Bar(y=train_df['language'],
           x=train_df['count'],
           orientation='h',
           name='Training Data'),
    row=2, col=1
)

# Update layout
fig.update_layout(height=800,  # Reduced height since we have fewer bars
                 showlegend=True,
                 title_text="Top 10 Languages Distribution Across All Models")

# Update axes
fig.update_xaxes(title_text="Number of Comments")
fig.update_yaxes(title_text="Language")

# Show the plot
fig.show()

# Print summary statistics
print("\nTop 10 Languages in Test Data:")
total_test_comments = test_df['count'].sum()
for _, row in test_df.iterrows():
    percentage = (row['count'] / total_test_comments * 100).round(2)
    print(f"{row['language']}: {row['count']} comments ({percentage}%)")

print("\nTop 10 Languages in Training Data:")
total_train_comments = train_df['count'].sum()
for _, row in train_df.iterrows():
    percentage = (row['count'] / total_train_comments * 100).round(2)
    print(f"{row['language']}: {row['count']} comments ({percentage}%)")
```
